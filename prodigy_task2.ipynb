```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
pclass_names = {
    1:"First class",
    2:"Second class",
    3:"Third class"
}
df['pclass name']=df['pclass'].map(pclass_names)
df.head()
```

```python
df['Family']=df['sibsp']+df["parch"]
df.head()
```

```python
df.describe().T
```

```python
cont_val = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare', 'who']]
sns.set_style('darkgrid')
sns.set_palette('pastel')
sns.heatmap(cont_val.corr(),cmap='Greens',annot=True)
```

```python
sns.countplot(data=df,x="pclass name" , hue='sex')
```

```python
sns.set_style('white')
sns.displot(data=df,x="age",bins=15,palette='dark',hue='survived',hue_order=[1,0])
```

```python
survived=df[df["survived"]==1]
ax=sns.countplot(data=df,x="pclass name",hue='sex',palette='dark')
ax=sns.countplot(data=survived,x="pclass name",hue='sex')
plt.legend(labels=['Male','Female','survived Male','survived Female'])
```
